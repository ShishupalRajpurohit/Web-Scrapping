```python
# so now we can use the requests for extracting the data with header as headers data
response = requests.get(url = 'https://www.nseindia.com/api/chart-databyindex?index=NIFTY%2050&indices=true', headers = header)
```

```python
response # <Response [200]> means the data successfully extracted (scrapped)
```

```python
response.json() 
```

```python
response.json()['grapthData'] # we only need graothData 
```

```python
# creating the dataframe and naming columns as timestamp and price
Nifty50 = pd.DataFrame(response.json()['grapthData'],columns=['timestamp','price'])
Nifty50
```

```python
# timestamp is in Epoch format so converting in into the datetime format
Nifty50['timestamp'] = pd.to_datetime(Nifty50['timestamp'],unit='ms')
```

```python
# plotting the graph and we can see its as it is as NSE site graph
Nifty50.plot(x='timestamp',y='price')
```

```python
# Saving it to the dataframe
Nifty50.to_csv('Nifty50.csv', index=False)
```
